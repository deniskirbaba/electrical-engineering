```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.array([2, 5, 20, 50, 200, 500, 2000, 5000, 20000, 50000])
```

```python
v_in = np.array([10, 10, 10, 10, 10, 10, 10, 10, 10, 10])

v_out = np.array([9.999, 9.994, 19.8/2, 19/2, 12.4/2, 5.99/2, 1.57/2, 0.623/2, 0.156/2, 0.0621/2])
v_out_2 = np.array([0.254/2, 0.636/2, 2.5/2, 6.02/2, 15.5/2, 18.7/2, 19.9/2, 19.9/2, 19.9/2, 19.9/2])
v_out_3 = np.array([0.254, 0.627, 2.37, 4.85, 6.54, 4.82, 1.52, 0.621, 0.157, 0.062]) / 2
```

```python
k = v_out/v_in
k_2 = v_out_2/v_in
k_3 = v_out_3/v_in
```

```python
c = 10**(-6)
r = 1000
```

```python
more_f = np.logspace(0, 5, 1000)
om_tau = more_f*2*np.pi*r*c

more_k = 1/np.sqrt(np.power(om_tau, 2) + 1)
more_k_2 = om_tau/np.sqrt(np.power(om_tau, 2) + 1)
more_k_3 = 1/np.sqrt(9 + np.power(om_tau - 1/om_tau, 2))
```

```python
cutting_point = np.array([1/(2*np.pi*c*r), 1/np.sqrt(2)])
cutting_point_2 = np.array([1/(2*np.pi*c*r), 1/np.sqrt(2)])
cutting_point_3 = np.array([1/(2*np.pi*c*r), 1/3])
```

```python
plt.figure(figsize=((16, 8)))
plt.xscale('log')
plt.grid()
plt.title('Amplitude frequency response')
plt.xlabel('F, Hz')
plt.ylabel('K')

plt.scatter(f, k, color='green', label='experimental')
plt.plot(more_f, more_k, color='blue', label='analytical')
plt.plot(cutting_point[0], cutting_point[1], '*', markersize=9, label = 'cutting dot',color='black')
plt.axvline(x = cutting_point[0], color = 'cyan', linestyle='--', label = f'f_c = {round(cutting_point[0], 2)} Hz')
plt.axhline(y = cutting_point[1], color = 'magenta', linestyle='--', label= f'k = {round(cutting_point[1], 2)}')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=((16, 8)))
plt.xscale('log')
plt.grid()
plt.title('Amplitude frequency response')
plt.xlabel('F, Hz')
plt.ylabel('K')

plt.scatter(f, k_2, color='green', label='experimental')
plt.plot(more_f, more_k_2, color='blue', label='analytical')
plt.plot(cutting_point[0], cutting_point[1], '*', markersize=9, label = 'cutting dot',color='black')
plt.axvline(x = cutting_point[0], color = 'cyan', linestyle='--', label = f'f_c = {round(cutting_point[0], 2)} Hz')
plt.axhline(y = cutting_point[1], color = 'magenta', linestyle='--', label= f'k = {round(cutting_point[1], 2)}')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=((16, 8)))
plt.xscale('log')
plt.grid()
plt.title('Amplitude frequency response')
plt.xlabel('F, Hz')
plt.ylabel('K')

plt.scatter(f, k_3, color='green', label='experimental')
plt.plot(more_f, more_k_3, color='blue', label='analytical')
plt.plot(cutting_point_3[0], cutting_point_3[1], '*', markersize=9, label = 'cutting dot',color='black')
plt.axvline(x = cutting_point_3[0], color = 'cyan', linestyle='--', label = f'f_c = {round(cutting_point_3[0], 2)} Hz')
plt.axhline(y = cutting_point_3[1], color = 'magenta', linestyle='--', label= f'k = {round(cutting_point_3[1], 2)}')

plt.legend()
plt.show()
```

```python
more_phi = np.arctan(-om_tau)
more_phi_2 = np.arctan(1/om_tau)
more_phi_3 = np.arctan(-1/3*(om_tau - 1/om_tau))
```

```python
cutting_phi = np.arctan(-2*np.pi*cutting_point[0]*c*r)
cutting_phi_2 = np.arctan(1/(2*np.pi*cutting_point[0]*c*r))
cutting_phi_3 = np.arctan(-1/3*(2*np.pi*cutting_point_3[0]*c*r - 1/(2*np.pi*cutting_point_3[0]*c*r)))
```

```python
plt.figure(figsize=((16, 8)))
plt.xscale('log')
plt.grid()
plt.title('Phase frequency response')
plt.xlabel('F, Hz')
plt.ylabel('phi')

plt.plot(more_f, more_phi, color='blue', label='analytical')
plt.plot(cutting_point[0], cutting_phi, '*', markersize=9, label = 'cutting dot',color='black')
plt.axvline(x = cutting_point[0], color = 'cyan', linestyle='--', label = f'f_c = {round(cutting_point[0], 2)} Hz')
plt.axhline(y = cutting_phi, color = 'magenta', linestyle='--', label= f'phi = {round(cutting_phi, 2)}')

plt.legend()
plt.show()
```
